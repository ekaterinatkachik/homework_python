```python
n, m, s, f = map(int, input().split())
graph = [[] for i in range(n)]
for i in range(m):
    u, v, w = map(int, input().split())
    graph[u].append((v, w))
    graph[v].append((u, w))

INF = 10**10
dist = [INF] * n
num_vertices = [0] * n
used = [False] * n

dist[s] = 0
num_vertices[s] = 1

for j in range(n):
    v = -1
    best = INF
    for i in range(n):
        if not used[i] and dist[i] < best:
            best = dist[i]
            v = i
    if v == -1:
        break
    used[v] = True
    for to, w in graph[v]:
        if not used[to]:
            new_dist = dist[v] + w
            if new_dist < dist[to]:
                dist[to] = new_dist
                num_vertices[to] = num_vertices[v] + 1

print(num_vertices[f])
```

Задача 2

```python
input()
s = input().strip()
s = s.replace('{', '[').replace('}', ']')
graph_array = eval(s)


def is_cyclic(graph_array: list[list[int]]) -> bool:
    used = [False] * len(graph_array)
    res = False

    def dfs(v, p=-1):
        nonlocal res
        used[v] = True
        for u in graph_array[v]:
            if not used[u]:
                dfs(u, v)
            elif u != p:
                res = True
                return

    for i in range(len(graph_array)):
        if not used[i]:
            dfs(i, -1)
            if res:
                break

    return res


if is_cyclic(graph_array):
    print('YES')
else:
    print('NO')
```

Задача 3

```python
n = int(input())
starts = list(map(int, input().split()))
ends = list(map(int, input().split()))

meetings = []
for i in range(n):
    meetings.append((starts[i], ends[i]))

meetings.sort(key=lambda x: x[1])

count = 0
last_end = -1

for start, end in meetings:
    if start > last_end:
        count += 1
        last_end = end

print(count)
```

Задача 4

```python
import heapq
from collections import defaultdict

parts = input().split()
V = int(parts[0])
E = int(parts[1])
start = parts[2]
end = parts[3]

graph = defaultdict(list)

for i in range(E):
    line = input().split()
    a = line[0]
    b = line[1]
    t = int(line[2])
    graph[a].append((b, t))
    graph[b].append((a, t))


def dijkstra(graph, start, end):
    dist = defaultdict(lambda: 10**10)
    parent = {}
    dist[start] = 0
    heap = [(0, start)]
    visited = set()

    while heap:
        d, u = heapq.heappop(heap)
        if u in visited:
            continue
        visited.add(u)
        if u == end:
            break
        for v, w in graph[u]:
            if v not in visited:
                nd = d + w
                if nd < dist[v]:
                    dist[v] = nd
                    parent[v] = u
                    heapq.heappush(heap, (nd, v))


    path = []
    cur = end
    while cur != start:
        path.append(cur)
        cur = parent[cur]
    path.append(start)
    path.reverse()
    return path, dist[end]


path, min_time = dijkstra(graph, start, end)

print(min_time)

f = open("otvet.txt", "w", encoding="utf-8")
if path:
    arrow_path = "→".join(path)
    f.write(f"Кратчайший путь от '{start}' до '{end}':\n")
    f.write(arrow_path + "\n")
    f.write(f"\nМинимальное время: {min_time} минут\n")
else:
    f.write("Путь не найден.\n")
f.close()
```

Задача 5

```python
n, m = map(int, input().split())
graph = [[] for i in range(n)]

for i in range(m):
    u, v = map(int, input().split())
    graph[u].append(v)
    graph[v].append(u)

dist = [-1] * n
dist[0] = 0

queue = [0]

while queue:
    current = queue.pop(0)
    for neighbor in graph[current]:
        if dist[neighbor] == -1:
            dist[neighbor] = dist[current] + 1
            queue.append(neighbor)

for d in dist:
    print(d)
```

Задача 6

```python
n = int(input())
m = int(input())
graph = [[] for i in range(n)]

for i in range(m):
    u, v = map(int, input().split())
    graph[u].append(v)
    graph[v].append(u)

visited = [False] * n
components = 0


def dfs(v):
    visited[v] = True
    for neighbor in graph[v]:
        if not visited[neighbor]:
            dfs(neighbor)


for i in range(n):
    if not visited[i]:
        components += 1
        dfs(i)

print(components)
```
