```python
from collections import defaultdict


def find_social_clusters(n, relations):
    graph = defaultdict(list)
    vertices = set()
    for i in range(n):
        vertices.add(i)

    for u, v in relations:
        graph[u].append(v)
        vertices.add(u)
        vertices.add(v)

    for i in range(n):
        if i not in graph:
            graph[i] = []

    visited = {v: False for v in vertices}
    finishing_order = []

    def dfs1(v):
        visited[v] = True
        for w in graph[v]:
            if not visited[w]:
                dfs1(w)
        finishing_order.append(v)

    for v in vertices:
        if not visited[v]:
            dfs1(v)

    transposed = defaultdict(list)
    for u in graph:
        for v in graph[u]:
            transposed[v].append(u)
    for v in vertices:
        if v not in transposed:
            transposed[v] = []

    visited = {v: False for v in vertices}
    sccs = []

    def dfs2(v, component):
        visited[v] = True
        component.append(v)
        for w in transposed[v]:
            if not visited[w]:
                dfs2(w, component)

    for v in reversed(finishing_order):
        if not visited[v]:
            comp = []
            dfs2(v, comp)
            sccs.append(comp)

    for comp in sccs:
        comp.sort()
    sccs.sort(key=lambda x: -len(x))

    return sccs


n1 = 6
relations1 = [(0, 1), (1, 2), (2, 0), (3, 4), (4, 5), (5, 3), (2, 3)]
result1 = find_social_clusters(n1, relations1)
print(result1)

n2 = 4
relations2 = [(0, 1), (1, 0)]
result2 = find_social_clusters(n2, relations2)
print(result2)
```

```python
from collections import deque


def kon(n, start_r, start_c):
    dist = []
    for i in range(n):
        a = []
        for j in range(n):
            a.append(-1)
        dist.append(a)

    moves = [[-2, -1], [-2, 1], [-1, -2], [-1, 2], [1, -2], [1, 2], [2, -1], [2,  1]]

    q = deque()
    q.append((start_r, start_c))
    dist[start_r][start_c] = 0

    while q:
        r, c = q.popleft()
        for i in range(8):
            dr = moves[i][0]
            dc = moves[i][1]
            nr = r + dr
            nc = c + dc
            if 0 <= nr < n and 0 <= nc < n and dist[nr][nc] == -1:
                dist[nr][nc] = dist[r][c] + 1
                q.append((nr, nc))

    return dist


n = 5
start_r = 0
start_c = 0
result = kon(n, start_r, start_c)
print(result[3][3])
```

```python
def dijkstra_max(n, w, start):
    inf = 10**10
    dist = [inf] * n
    used = [False] * n
    dist[start] = 0

    min_val = 0
    min_vertex = start

    while min_val < inf:
        i = min_vertex
        used[i] = True

        for j in range(n):
            if not used[j] and w[i][j] > 0:
                new_max = max(dist[i], w[i][j])
                if new_max < dist[j]:
                    dist[j] = new_max

        min_val = inf
        for j in range(n):
            if not used[j] and dist[j] < min_val:
                min_val = dist[j]
                min_vertex = j

    return dist


n = 3
w = [[0, 5, 0], [0, 0, 3], [0, 0, 0]]
start = 0
print(dijkstra_max(n, w, start))
```

```python
def dijkstra_product(n, w, start):
    inf = 10**10
    dist = [inf] * n
    used = [False] * n
    dist[start] = 1

    min_val = 1
    min_vertex = start

    while min_val < inf:
        i = min_vertex
        used[i] = True

        for j in range(n):
            if not used[j] and w[i][j] >= 1:
                new_prod = dist[i] * w[i][j]
                if new_prod < dist[j]:
                    dist[j] = new_prod

        min_val = inf
        for j in range(n):
            if not used[j] and dist[j] < min_val:
                min_val = dist[j]
                min_vertex = j

    return dist


n = 3
w = [[0, 2, 0], [0, 0, 3], [0, 0, 0]]
start = 0
print(dijkstra_product(n, w, start))
```

```python
def dijkstra_str(n, w, start):
    inf = "~" * 1000
    dist = [inf] * n
    used = [False] * n
    dist[start] = ""
    current_min = ""
    min_vertex = start

    for e in range(n):
        i = min_vertex
        if used[i]:
            break
        used[i] = True

        for j in range(n):
            if not used[j] and w[i][j] != "":
                new_str = dist[i] + w[i][j]
                if new_str < dist[j]:
                    dist[j] = new_str

        min_vertex = -1
        current_min = inf
        for j in range(n):
            if not used[j] and dist[j] < current_min:
                current_min = dist[j]
                min_vertex = j

        if min_vertex == -1:
            break

    return dist


n = 3
w = [["", "b", ""], ["", "", "a"], ["", "", ""]]
start = 0
print(dijkstra_str(n, w, start))
```

```python
def dijkstra_min(n, w, start):
    inf = -1
    dist = [inf] * n
    used = [False] * n
    dist[start] = 10**10

    max_val = 10**10
    min_vertex = start

    while max_val > 0:
        i = min_vertex
        used[i] = True

        for j in range(n):
            if not used[j] and w[i][j] > 0:
                new_min = min(dist[i], w[i][j])
                if new_min > dist[j]:
                    dist[j] = new_min

        max_val = inf
        next_vertex = -1
        for j in range(n):
            if not used[j] and dist[j] > max_val:
                max_val = dist[j]
                next_vertex = j

        if next_vertex == -1:
            break
        min_vertex = next_vertex

    return dist


n = 3
w = [[0, 5, 0], [0, 0, 3], [0, 0, 0]]
start = 0
print(dijkstra_min(n, w, start))
```

```python
def dijkstra_colour(n, w, colour, start):
    inf = (10**10, 10**10)
    dist = [inf] * n
    used = [False] * n
    dist[start] = (0, 0)

    current_min = (0, 0)
    min_vertex = start

    for _ in range(n):
        i = min_vertex
        if used[i]:
            break
        used[i] = True

        for j in range(n):
            if not used[j] and w[i][j] > 0:
                change = 1 if colour[i] != colour[j] else 0
                new_changes = dist[i][0] + change
                new_length = dist[i][1] + w[i][j]
                new_pair = (new_changes, new_length)

                if new_pair < dist[j]:
                    dist[j] = new_pair

        min_vertex = -1
        current_min = inf
        for j in range(n):
            if not used[j] and dist[j] < current_min:
                current_min = dist[j]
                min_vertex = j

        if min_vertex == -1:
            break

    return dist


n = 4
w = [[0, 2, 0, 0], [0, 0, 3, 0], [0, 0, 0, 1], [0, 0, 0, 0]]
colour = [0, 1, 1, 0]
start = 0
print(dijkstra_colour(n, w, colour, start))
```
