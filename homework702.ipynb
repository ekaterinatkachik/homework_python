```python
n = int(input())
vershini = set()
graph = {}

for i in range(n):
    ver1, ver2 = map(int, input().split())
    vershini.add(ver1)
    vershini.add(ver2)

    if ver1 not in graph:
        graph[ver1] = []
    if ver2 not in graph:
        graph[ver2] = []

    graph[ver1].append(ver2)
    graph[ver2].append(ver1)

vershini = list(vershini)
if len(vershini) == 0:
    print(True)
else:
    visited = set()
    stack = [vershini[0]]

    while stack:
        tekushiy = stack.pop()
        if tekushiy not in visited:
            visited.add(tekushiy)
            for neib in graph[tekushiy]:
                if neib not in visited:
                    stack.append(neib)

    print(len(visited) == len(vershini))
```

```python
n = int(input())
stroka = input()
stroka = stroka.replace('[', ' ').replace(']', ' ').replace('(', ' ').replace(')', ' ').replace(',', ' ')
alll = list(map(int, stroka.split()))

rebra = []
for i in range(0, len(alll), 2):
    start_uzel = alll[i]
    end_uzel = alll[i + 1]
    rebra.append((start_uzel, end_uzel))

start, end = map(int, input().split())

graph = {}

for a, b in rebra:
    if a not in graph:
        graph[a] = []
    graph[a].append(b)

if start not in graph:
    graph[start] = []
if end not in graph:
    graph[end] = []

visited = set()
stack = [start]
found = False

while stack:
    tekush = stack.pop()
    if tekush == end:
        found = True
        break
    if tekush not in visited:
        visited.add(tekush)
        for neighbor in graph[tekush]:
            if neighbor not in visited:
                stack.append(neighbor)

print(found)
```

```python
stroka = input()
stroka = stroka.replace('{', ' ').replace('}', ' ').replace(':', ' ').replace(',', ' ').replace("'", ' ')
aero = stroka.split()

fly = {}
for i in range(0, len(aero), 2):
    if i + 1 < len(aero):
        fromm = aero[i]
        too = aero[i + 1]
        fly[fromm] = too

prilet = set(fly.values())
start = None
for airoport in fly.keys():
    if airoport not in prilet:
        start = airoport
        break

trip = []
tekushiy = start
while tekushiy is not None:
    trip.append(tekushiy)
    tekushiy = fly.get(tekushiy)

print(trip)
```

```python
n = int(input())
graph = {}
vershini = set()

for i in range(n):
    a, b = map(int, input().split())
    vershini.add(a)
    vershini.add(b)

    if a not in graph:
        graph[a] = []
    if b not in graph:
        graph[b] = []

    graph[a].append(b)
    graph[b].append(a)

visited = set()
est_tsikl = False

for start in vershini:
    if start in visited:
        continue

    stack = [(start, -1)]
    visited.add(start)

    while stack and not est_tsikl:
        tekushiy, parent = stack.pop()

        for neighbor in graph[tekushiy]:
            if neighbor == parent:
                continue
            if neighbor in visited:
                est_tsikl = True
                break
            visited.add(neighbor)
            stack.append((neighbor, tekushiy))

    if est_tsikl:
        break

print(est_tsikl)
```

```python
s = input()
s = s.replace('[', ' ').replace(']', ' ').replace('(', ' ').replace(')', ' ').replace(',', ' ')

chisla = list(map(int, s.split()))

rebra = []
for i in range(0, len(chisla), 3):
    ot = chisla[i]
    kuda = chisla[i+1]
    ves = chisla[i+2]
    rebra.append((ot, kuda, ves))

start, end = map(int, input().split())

vershiny = set()
for ot, kuda, ves in rebra:
    vershiny.add(ot)
    vershiny.add(kuda)

graf = {}
for v in vershiny:
    graf[v] = []

for ot, kuda, ves in rebra:
    graf[ot].append((kuda, ves))

rastoyaniya = {}
for v in vershiny:
    rastoyaniya[v] = 1000000000
rastoyaniya[start] = 0

poseshcheno = set()

while True:
    luchshaya = None
    min_rast = 1000000000

    for v in vershiny:
        if v not in poseshcheno and rastoyaniya[v] < min_rast:
            min_rast = rastoyaniya[v]
            luchshaya = v

    if luchshaya is None:
        break

    if luchshaya == end:
        break

    poseshcheno.add(luchshaya)

    for soseda, ves in graf[luchshaya]:
        novoe = rastoyaniya[luchshaya] + ves
        if novoe < rastoyaniya[soseda]:
            rastoyaniya[soseda] = novoe

print(rastoyaniya[end])
```

```python
s = input()
s = s.replace('[', ' ').replace(']', ' ').replace('(', ' ').replace(')', ' ').replace(',', ' ').replace('"', ' ').replace("'", ' ')
words = s.split()

if len(words) == 0:
    print(False)
else:
    in_deg = {}
    out_deg = {}
    vershini = set()
    rebra = []

    for word in words:
        a = word[0]
        b = word[-1]

        if a in out_deg:
            out_deg[a] += 1
        else:
            out_deg[a] = 1

        if b in in_deg:
            in_deg[b] += 1
        else:
            in_deg[b] = 1

        vershini.add(a)
        vershini.add(b)
        rebra.append((a, b))

    ok = True
    for v in vershini:
        d_in = in_deg.get(v, 0)
        d_out = out_deg.get(v, 0)
        if d_in != d_out:
            ok = False
            break

    if not ok:
        print(False)
    else:
        graph = {}
        for v in vershini:
            graph[v] = []

        for x, y in rebra:
            graph[x].append(y)
            graph[y].append(x)

        spisok_vershin = list(vershini)
        pervaya = spisok_vershin[0]

        visited = set()
        stack = [pervaya]

        while stack:
            u = stack.pop()
            if u not in visited:
                visited.add(u)
                for w in graph[u]:
                    if w not in visited:
                        stack.append(w)

        if len(visited) == len(vershini):
            print(True)
        else:
            print(False)
```
