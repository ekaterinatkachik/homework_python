Задача 1

```python
import math
n = int(input())
koord = []
for i in range(n):
    x, y = map(int, input().split())
    koord.append((x, y))


def distance(i, j):
    x1, y1 = koord[i]
    x2, y2 = koord[j]
    return math.sqrt((x1-x2)**2 + (y1-y2)**2)


dist = [[0.0] * n for i in range(n)]


for i in range(n):
    for j in range(n):
        dist[i][j] = distance(i, j)


for k in range(n):
    for i in range(n):
        for j in range(n):
            cherez_k = max(dist[i][k], dist[k][j])
            dist[i][j] = min(dist[i][j], cherez_k)


print(f"{dist[0][1]:.3f}")
```

Задача 2

```python
c = int(input())

for i in range(c):
    n, m = map(int, input().split())

    cherv = []
    for j in range(m):
        cherv.append(list(map(int, input().split())))

    dist = [10**9] * n
    dist[0] = 0

    cycle = False

    for i in range(n):
        ism = False
        for u, v, t in cherv:
            if dist[u] + t < dist[v]:
                dist[v] = dist[u] + t
                ism = True

        if i == n - 1 and ism:
            cycle = True

    if cycle:
        print("возможно")
    else:
        print("не возможно")
```
